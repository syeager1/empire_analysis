```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
empire_raw = open('empire2.txt', 'r') #import the file for CountVectorizer 
```

```python
#adapted from the workshop's part1
example_vectorizer = CountVectorizer() #initialize your count vectorizer
example_vectorizer.fit(empire_raw) #documents much be a vector of strings(individual documents)
print("Vocabulary size:", len(example_vectorizer.vocabulary_))
example_vectorizer.vocabulary_  #We can get the unique vocabulary set and its corresponding index here
```

```python
#converts to file to a list of strings
empire_list = list(empire_raw)
for word in empire_list:
    word.decode('utf-8')
```

```python
counts = example_vectorizer.transform(empire_list)
print(type(counts))
print(counts)
print("   ↑  ↑         ↑\n  doc word_id count")
```

```python
doc = 0  # first document
for word, word_id in example_vectorizer.vocabulary_.items():
    print(word, ":", counts[doc, word_id])
```
